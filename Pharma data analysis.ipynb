## What are the total sales quantities for each drug category (ATC code)?

```python
fig, ax = plt.subplots()



bar_labels = ['purple', 'blue', 'green']
bar_colors = ['tab:purple', 'tab:blue', 'tab:green']
bar_container=ax.bar(["RCode","NCode","MCode"],daily[["RCode","MCode","NCode"]].sum(),label=bar_labels, color=bar_colors)
ax.set_title('Total sales for each drug category (ATC code)')
ax.bar_label(bar_container)


plt.show()
```

## Which three drugs have the highest sales in January 2015, July 2016, September 2017.

```python
monthly.head()
```

```python
monthly["datum"]=pd.to_datetime(monthly["datum"])
```

```python
sep2017=monthly[(monthly["datum"].dt.year==2017) & (monthly["datum"].dt.month==9)].transpose().reset_index().drop(index=0).sort_values(by=[44],ascending=False)
sep2017.drop(sep2017.tail(5).index, inplace=True)
sep2017["index"]=sep2017["index"].apply(lambda i : i + " Sep2017")
sep2017
```

```python
jul2016=monthly[(monthly["datum"].dt.year==2016) & (monthly["datum"].dt.month==7)].transpose().reset_index().drop(index=0).sort_values(by=[30],ascending=False)
jul2016.drop(jul2016.tail(5).index, inplace=True)
jul2016["index"]=jul2016["index"].apply(lambda i : i + " Jul2016")
jul2016
```

```python
jan2015=monthly[(monthly["datum"].dt.year==2015) & (monthly["datum"].dt.month==1)].transpose().reset_index().drop(index=0).sort_values(by=[12],ascending=False)
jan2015.drop(jan2015.tail(5).index, inplace=True)
jan2015["index"]=jan2015["index"].apply(lambda i : i + " Jan2015")
jan2015
```

```python
fig, ax = plt.subplots()



ax.bar(jan2015["index"],jan2015[12],width=0.8, color='b', align='center', label="Jan2015")
ax.bar(jul2016["index"],jul2016[30], width=0.8, color='g', align='center', label="Jul2016")
ax.bar(sep2017["index"],sep2017[44],width=0.8, color='r', align='center', label="Sep2017")
ax.set_title('Top 3 drugs on Jan 2015, Jul 2016, Sep 2017')
plt.setp(ax.get_xticklabels(), rotation=40, horizontalalignment='right')
plt.legend()
ax.bar_label(ax.bar(jan2015["index"],jan2015[12],width=0.8, color='b', align='center', label="Jan2015"))
ax.bar_label( ax.bar(jul2016["index"],jul2016[30], width=0.8, color='g', align='center', label="Jul2016") )
ax.bar_label(ax.bar(sep2017["index"],sep2017[44],width=0.8, color='r', align='center', label="Sep2017"))

plt.show()
```

## Which drug category has the highest average daily sales?

```python
hourly.head()
```

```python
hourly["MCode"]=hourly["M01AB"]+hourly["M01AE"]
hourly[["MCode","M01AB","M01AE"]]
```

```python
hourly["NCode"]=hourly["N02BA"]+daily["N02BE"]+hourly["N05B"]+hourly["N05C"]
hourly[["NCode","N02BA","N02BE","N05B","N05C"]]
```

```python
hourly["RCode"]=hourly["R03"]+daily["R06"]
hourly[["RCode","R03","R06"]]
```

```python
hourly.isna().sum()
```

```python
hourly.fillna(0, inplace=True)
```

```python
hourly["day"]=pd.to_datetime(hourly["datum"].apply(lambda d:d.split(" ")[0]))
```

```python
category_avg=hourly.groupby('day').agg(lambda x: x.ne(0).sum()).mean()[13:].sort_values(ascending=False)
```

```python
fig, ax = plt.subplots()



bar_labels = ['b', 'g', 'r']
bar_colors = ['tab:blue', 'tab:green', 'tab:red']
bar_container=ax.bar(category_avg.index,category_avg,width=0.8, label=bar_labels, color=bar_colors)
ax.set_title('Drug category average daily sales')
plt.setp(ax.get_xticklabels(), rotation=40, horizontalalignment='right')

ax.bar_label(bar_container)

plt.show()
```

## Are respiratory drugs (R03) sold more during specific months

```python
R03_per_month=hourly[["Month","R03"]].groupby("Month").sum()
R03_per_month
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

bar_container=ax.bar(R03_per_month.index,R03_per_month["R03"],width=0.5)
ax.set_title('R03 sales per month')

plt.xticks(R03_per_month.index)
ax.bar_label(bar_container)

plt.show()
```
